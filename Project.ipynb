```python
import numpy as np
import matplotlib.pyplot as plt

# Remove ID_REF column
data = hip_expression.iloc[:,1:]

# Log2 transformation
log_data = np.log2(data + 1)

# Drop rows with NaN values before PCA
log_data = log_data.dropna()

# Flatten
values = log_data.values.flatten()

# Remove NaN values (this line is mostly redundant after dropna but kept for safety in flattening)
values = values[~np.isnan(values)]

plt.figure(figsize=(10,6))
plt.hist(values, bins=50, edgecolor='black')
plt.title("Histogram of Log2 Gene Expression")
plt.xlabel("Log2(Expression + 1)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(15,6))

log_data.boxplot(rot=90)

plt.title("Boxplot of Log2 Gene Expression")
plt.xlabel("Samples")
plt.ylabel("Log2(Expression + 1)")

plt.show()
```

```python
from scipy.stats import ttest_ind
import numpy as np
import pandas as pd

# Separate Control and Alzheimer's samples

control_samples = hip_samples[
    hip_samples["Sample_Name"].str.contains("control", case=False)
]["GSM_ID"].tolist()

ad_samples = hip_samples[
    hip_samples["Sample_Name"].str.contains("affected", case=False)
]["GSM_ID"].tolist()

print(control_samples)
print(ad_samples)
```

```python
results = []

for i in range(len(hip_expression)):

    gene = hip_expression.iloc[i, 0]

    control = hip_expression.loc[i, control_samples].values.astype(float)

    ad = hip_expression.loc[i, ad_samples].values.astype(float)

    log2FC = np.log2(ad.mean() + 1) - np.log2(control.mean() + 1)

    _, pvalue = ttest_ind(ad, control, equal_var=False)

    results.append([gene, log2FC, pvalue])

results_df = pd.DataFrame(
    results,
    columns=["Gene", "log2FC", "pvalue"]
)

results_df.head()
```

```python
results_df["minusLog10P"] = -np.log10(results_df["pvalue"])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))

plt.scatter(
    results_df["log2FC"],
    results_df["minusLog10P"],
    s=8,
    alpha=0.5
)

plt.axhline(-np.log10(0.05),
            color="red",
            linestyle="--")

plt.axvline(1,
            color="green",
            linestyle="--")

plt.axvline(-1,
            color="green",
            linestyle="--")

plt.xlabel("Log2 Fold Change")
plt.ylabel("-Log10(p-value)")
plt.title("Volcano Plot")

plt.show()
```

```python
significant = results_df[
    (results_df["pvalue"] < 0.05) &
    (abs(results_df["log2FC"]) > 1)
]

print("Number of Significant Genes:", len(significant))

significant.head()
```

```python
significant.to_excel(
    "Significant_Genes.xlsx",
    index=False
)
```

```python
top20 = significant.sort_values(
    by="pvalue"
).head(20)

top20
```

```python
plt.figure(figsize=(10,8))

# Not significant
plt.scatter(
    results_df[
        (results_df["pvalue"] >= 0.05) |
        (abs(results_df["log2FC"]) <= 1)
    ]["log2FC"],
    results_df[
        (results_df["pvalue"] >= 0.05) |
        (abs(results_df["log2FC"]) <= 1)
    ]["minusLog10P"],
    color="lightgray",
    s=8
)

# Upregulated
plt.scatter(
    significant[significant["log2FC"] > 1]["log2FC"],
    significant[significant["log2FC"] > 1]["minusLog10P"],
    color="red",
    s=8,
    label="Upregulated"
)

# Downregulated
plt.scatter(
    significant[significant["log2FC"] < -1]["log2FC"],
    significant[significant["log2FC"] < -1]["minusLog10P"],
    color="blue",
    s=8,
    label="Downregulated"
)

plt.axhline(-np.log10(0.05),
            color='black',
            linestyle='--')

plt.axvline(1,
            color='black',
            linestyle='--')

plt.axvline(-1,
            color='black',
            linestyle='--')

plt.xlabel("Log2 Fold Change")
plt.ylabel("-Log10(p-value)")
plt.title("Volcano Plot")
plt.legend()

plt.show()
```

```python
X = log_data.T

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)

principal_components = pca.fit_transform(X_scaled)

print(principal_components.shape)
```

```python
labels = []

for sample in hip_samples["Sample_Name"]:

    if "control" in sample.lower():
        labels.append("Control")
    else:
        labels.append("Alzheimer")

plt.figure(figsize=(8,6))

for label in ["Control", "Alzheimer"]:

    idx = [i for i, l in enumerate(labels) if l == label]

    plt.scatter(
        principal_components[idx, 0],
        principal_components[idx, 1],
        s=80,
        label=label
    )

plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]*100:.2f}% Variance)")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]*100:.2f}% Variance)")
plt.title("PCA Plot of Hippocampus Samples")

plt.legend()
plt.grid(alpha=0.3)

plt.show()
```

```python
plt.figure(figsize=(8,6))

colors = {
    "Control": "royalblue",
    "Alzheimer": "crimson"
}

for label in ["Control", "Alzheimer"]:

    idx = [i for i, l in enumerate(labels) if l == label]

    plt.scatter(
        principal_components[idx,0],
        principal_components[idx,1],
        s=90,
        c=colors[label],
        edgecolors="black",
        linewidth=0.6,
        label=label
    )

plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]*100:.2f}% Variance)")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]*100:.2f}% Variance)")
plt.title("PCA of Alzheimer's and Control Hippocampus Samples")

plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()

plt.show()
```
